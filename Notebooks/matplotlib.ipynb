This notebook is a slight variation on http://matplotlib.org/users/pyplot_tutorial.html.

```python
import numpy as np
import scipy.stats as scs
import pandas as pd
import matplotlib.pylab as plt
```

```python
plt.plot([1, 2, 3, 4])
plt.xlabel('some numbers')
plt.ylabel('some other numbers')
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro')
plt.axis([0, 6, 0, 20])
plt.show()
```

```python
t = np.arange(0., 5., .2)
plt.plot(t, t, 'r--', t, t**2, 'bs', t, t**3, 'g^')
plt.show()
```

```python
lines = plt.plot(t, t**3, linewidth=2, color='red')
plt.show()
```

```python
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)

t1 = np.arange(.0, 5, .1)
t2 = np.arange(.0, 5, .02)

plt.figure(1, figsize=(8, 4))
plt.subplot(121)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k')

plt.subplot(122)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1)
axes[0].plot(t1, f(t1), 'bo', t2, f(t2), 'k')
axes[1].plot(t2, np.cos(2*np.pi*t2), 'r--')
plt.show()
```

```python
x1 = np.linspace(0, 5)
x2 = np.linspace(0, 2)

y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.subplot(2, 1, 1)
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2 subplots')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.ylabel('Undamped')

plt.show()
```

```python
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, normed=1,
                            facecolor='g', alpha=0.75, lw=0)
plt.plot(bins, scs.norm.pdf(bins, mu, sigma), lw=2, c='b', ls='--')
plt.xlabel('Smarts')
plt.ylabel('Probability')
plt.title('Histogram of IQ')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.axis([40, 160, 0, 0.03])
plt.grid()
plt.show()
```

```python
t = np.arange(.0, 5., .01)
s = np.cos(2 * np.pi * t)

plt.plot(t, s, lw=2)
plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
             arrowprops=dict(facecolor='black', shrink=.05))
plt.ylim(-2,2)
plt.show()
```

```python
with plt.xkcd():
    plt.plot(t, s, lw=2)
    plt.annotate('local max', xy=(2, 1), xytext=(3, 1.5),
                 arrowprops=dict(facecolor='black', shrink=0.05))
    plt.ylim(-2,2)
    plt.show()
```

```python
n_groups = 5

means_men = (20, 35, 30, 35, 27)
std_men = (2, 3, 4, 1, 2)

means_women = (25, 32, 34, 20, 25)
std_women = (3, 5, 2, 3, 3)

groups = ('A', 'B', 'C', 'D', 'E')

fig, ax = plt.subplots()

index = np.arange(n_groups)
bar_width = 0.35

opacity = 0.4
error_config = {'ecolor': '0.3'}

rects1 = plt.bar(index, means_men, bar_width,
                 alpha=opacity,
                 color='b',
                 yerr=std_men,
                 error_kw=error_config,
                 label='Men')

rects2 = plt.bar(index + bar_width, means_women, bar_width,
                 alpha=opacity,
                 color='r',
                 yerr=std_women,
                 error_kw=error_config,
                 label='Women')

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(index + bar_width, groups)
plt.legend()

plt.tight_layout()
plt.show()
```
