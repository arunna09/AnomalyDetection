```python
main_signal = np.random.normal(0.5,5,10000) + np.random.normal(1.5,15,10000)
main_signal.shape
```

```python
plt.hist(main_signal)
plt.show()
```

```python
def estimate_gaussian(X):  
    mu = X.mean(axis=0)
    sigma = X.var(axis=0)

    return mu, sigma
```

```python
mu1, sigma1 = estimate_gaussian(main_signal)
mu1, sigma1
```

```python
m = np.random.normal(mu1, sigma1, 10000)
plt.hist(m)
plt.show()
```

```python
count, bins, ignored = plt.hist(m, 30, normed=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - 0.1)**2 / (2 * sigma**2) ),linewidth=2, color='r')
plt.show()
```

```python
mu_noise, sigm_noise = estimate_gaussian(noise)
mu_noise, sigm_noise
```

```python
add.outlier.to.vector <- function(vector, amount) {
  cells.to.modify <- sample(1:length(vector), amount, replace=F)
  mean.val <- mean(vector)
  sd.val <- sd(vector)
  min.val <- mean.val - 3 * sd.val
  max.val <- mean.val + 3 * sd.val 
  vector[cells.to.modify] <- runif(amount, min=min.val, max=max.val)
  return(vector)
}
add.outlier.to.data.frame <- function (temp, amount){
  for (i in 1:ncol(temp)) {
    temp[,i] <- add.outlier.to.vector(temp[,i], amount)
  }
  return (temp)
} 

data <- data.frame(
  a=c(1,2,3,4),
  b=c(7,8,9,10)
)
add.outlier.to.data.frame(data, 2)
```
